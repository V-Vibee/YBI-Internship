**Defining X Features  & Y Target Variable**

```python
X= df.drop(['fetal_health'], axis=1)
y= df['fetal_health']
```

```python
df.columns
```

**Importing Libraries for Machine Learning/ Model Building**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import make_pipeline, Pipeline
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
```

**Splitting the data**

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.8, random_state=2529)
```

**Making Pipelines**

```python
pipe_knn= Pipeline([("scale1", StandardScaler()),
                  ('knn', KNeighborsClassifier())])

pipe_dt= Pipeline([("scale2",StandardScaler()),
                     ("decision_tree",DecisionTreeClassifier())])

pipe_gbc= Pipeline([("scale3",StandardScaler()),
                     ("gbc",GradientBoostingClassifier())])


pipe_rfc= Pipeline([("scale4",StandardScaler()),
                     ("rfc",RandomForestClassifier())])


```

```python
pipelines= [pipe_knn,pipe_dt, pipe_gbc, pipe_rfc]
```

**Creating a Pipe Dict so that we can fit these pipe easily**

```python
pipe_dict= {0:"KNN", 1: "DecisionTree", 2:'GradientBoost',3:"RandomForest"}
```

**Now fitting the pipelines**

```python
for pipe in pipelines:
  pipe.fit(X_train, y_train)
```

```python
pipe.fit(X_train, y_train)
```

```python
accuracy_scores = []

for i, model in enumerate(pipelines):
    cv_score = cross_val_score(model, X_train,y_train, cv=5)
    accuracy_scores.append(cv_score)
    print("%s: %f " % (pipe_dict[i], cv_score.mean()))
```

**Baseline Model**

```python
rf= RandomForestClassifier()

rf.fit(X_train, y_train)
```

```python
print(f"Raandom Forest Score: {round(rf.score(X_test, y_test), 2)}")

rf_prediction = rf.predict(X_test)
```

**Grid Search CV**

```python
params = {
    "n_estimators": [100,200,500],
    "max_depth": [3,6,9],
    "learning_rate": [0.01, 0.1,1],

}
```

```python
Gs = GridSearchCV(estimator=GradientBoostingClassifier(),
                                param_grid=params,
                                cv=5,
                                verbose=4,
                                refit="r2",
                                scoring= ["r2", "neg_root_mean_squared_error"],
                                return_train_score=True
                                )
```

```python
Gs.fit(X_train, y_train);
```

```python
print(Gs.best_estimator_)
```

```python
print(Gs.best_params_)
```

**Testing the Model**

```python
# Testing with the best parameters,

gbc = GradientBoostingClassifier( learning_rate=0.01,
                                  max_depth=3, n_estimators=500, random_state=2529)

gbc_model = gbc.fit(X_train, y_train)
y_pred_gbc = gbc_model.predict(X_test)

gbc_train_score = gbc_model.score(X_train, y_train)
gbc_test_score = gbc_model.score(X_test, y_test)
```

```python
pred_gbc = gbc.predict(X_test)
accuracy = accuracy_score(y_test, y_pred_gbc)
print(f" Testing Score of the model is {accuracy}")
```

```python
print(classification_report(y_test, y_pred_gbc))
```
