```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

**Import Dataset**

```python
df= pd.read_excel('https://github.com/V-Vibee/My-Projects/raw/main/Retail%20Datset/Online%20Retail.xlsx')
```

```python
df.head()
```

**Data Preparation**

```python
df.columns
```

```python
df['Description']= df['Description'].str.strip() #remove space
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True) #remove duplicates
df['InvoiceNo']= df['InvoiceNo'].astype('str') # convert to string
df= df[~df['InvoiceNo'].str.contains('C')] #Remove credit transactions
```

```python
df.head()
```

```python
df['Country'].value_counts()
```

```python
df.shape
```

**Taking out data of Germany**

```python
#Seperating tranactyion of germany

mybasket= (df[df['Country']=='Germany']
           .groupby(['InvoiceNo', 'Description'])['Quantity']
           .sum().unstack().reset_index().fillna(0)
           .set_index('InvoiceNo'))
```

```python
mybasket.head()
```

```python
#converting all postive values to 1

def my_encode_units(x):
  if x <=0:
    return 0
  if x >=1:
    return 1


my_basket_sets= mybasket.applymap(my_encode_units)
my_basket_sets.drop('POSTAGE', inplace=True, axis=1)
```

**Training the model**

```python
#generating frequent itemset

my_frequent_itemsets= apriori(my_basket_sets, min_support=0.07,use_colnames=True)

```

```python
#generating rules
my_rules= association_rules(my_frequent_itemsets, metric="lift", min_threshold=1)
```
