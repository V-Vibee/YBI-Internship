**SIMPLE Imputer**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Titanic.csv')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
# initiate with mean
si= SimpleImputer(missing_values= np.nan, strategy='mean')
```

```python
# impute with mean [depending on column values may use median, mode or constant]

df[['age']]= si.fit_transform(df[['age']])
```

```python
# initiate with mean and missing indicator

si = SimpleImputer(missing_values = np.nan, strategy= 'mean', add_indicator=True)
```

```python
# impute with mean with missing indicator

df[['fare','fare_indicator']] = si.fit_transform(df[['fare']])
```

```python
df.head()
```

```python
df.isna().sum()
```

**KNN Imputer**

```python
df= pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Titanic.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import KNNImputer
```

```python
knn = KNNImputer(missing_values = np.nan, weights ='distance')
```

```python
# impute with KNN [n_neighbors=5]

df[['age']] = knn.fit_transform(df[['age']])
```

```python
# initiate with missing indicator

knn = KNNImputer(missing_values = np.nan, weights ='distance', add_indicator=True)
```

```python
# impute with mean with missing indicator

df[['fare','fare_indicator']] = knn.fit_transform(df[['fare']])
```

**Iterative Imputer [default = RidgeRegressor()]**

```python
df= pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Titanic.csv')
```

```python
df.isna().sum()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
# initiate with default RidgeRegressor

ii = IterativeImputer(missing_values = np.nan)
```

```python
# impute with IterativeImputer [using default arguments]

df[['age']] = ii.fit_transform(df[['age']])
```

**Iterative Imputer [KNeighborsRegressor()]**

```python
df= pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Titanic.csv')
```

```python
df.isna().sum()
```

```python
# initiate with Random Forest Regressor
from sklearn.neighbors import KNeighborsRegressor

iiknn = IterativeImputer(missing_values = np.nan, estimator=KNeighborsRegressor())
```

```python
# impute with IterativeImputer [using KNN]
df[['age']] = iiknn.fit_transform(df[['age']])
```

```python
df.isna().sum()
```

**Iterative Imputer [RandomForestRegressor()]**

```python
df= pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Titanic.csv')
```

```python
df.isna().sum()
```

```python
# initiate with Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor

iirf = IterativeImputer(missing_values = np.nan, estimator=RandomForestRegressor())
```

```python
# impute with IterativeImputer [using Random Forest]
df[['age']] = iirf.fit_transform(df[['age']])
```
