**Logistic Regression**

```python
# Step 1 Import library

import pandas as pd
```

```python
# Step 2 Read datasets

df= pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Bank%20Churn%20Modelling.csv')
```

```python
df.head()
```

```python
# One hot encoding

df1= pd.get_dummies(df, columns=['Geography','Gender'])
```

```python
df1.head()
```

```python
df1.columns
```

```python
# Step 3 Define x & y

y= df1['Churn']

X= df1[[ 'CreditScore', 'Age', 'Tenure', 'Balance','Num Of Products', 'Has Credit Card', 'Is Active Member',
       'Estimated Salary', 'Geography_France', 'Geography_Germany','Geography_Spain', 'Gender_Female', 'Gender_Male']]
```

```python
# Step 4 Train test split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y, random_state=2529)
```

```python
# Step 5 Select a model

from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
```

```python
# step 6 Training the model

model.fit(X_train, y_train)
```

```python
# Step 7 Predict the model

y_pred= model.predict(X_test)
```

```python
## step 8 Evaluation

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
