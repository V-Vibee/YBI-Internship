***Supervised ML model: Prediction acccuracy needs to be High here***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

***Download the Datasets***

```python
df= pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/TelecomCustomerChurn.csv')
```

***Define y & X***

```python
y= df['Churn']
X= df.drop(['customerID','Churn'], axis=1)
```

```python
X
```

***We have to encode the categorical data***

***One-Hot Encoding***

```python
from sklearn.preprocessing import OrdinalEncoder
oe= OrdinalEncoder()
```

```python
 X=oe.fit_transform(X)
```

```python
X
```

***Scaling***

```python
y.value_counts()
```

***Here we have to convert Imbalance Dataset to balanced***

```python
from imblearn.over_sampling import RandomOverSampler
ros= RandomOverSampler()
```

```python
X,y= ros.fit_resample(X,y)
```

```python
y.value_counts()
```

#***Train Test Split***

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, random_state=2529)
```

***Select the model: Ensemble(Random Forest)***

```python
from sklearn.ensemble import RandomForestClassifier
rfc= RandomForestClassifier()
```

***Train the Model***

```python
rfc.fit(X_train, y_train)
```

***Model Prediction***

```python
y_pred= rfc.predict(X_test)
```

***Model Evaluation***

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

***Recall is very good for this Model stands AT 96% which says that this model is able to catch majority od customers***
