```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/IRIS.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
y= df['species']
X= df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width' ]]
```

```python
y= pd.get_dummies(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y, stratify=y, random_state=2529)
```

```python
from sklearn.preprocessing import StandardScaler
ss= StandardScaler()
```

```python
X_train= ss.fit_transform(X_train)
X_test= ss.transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

**Empty Wrapper**

```python
ann= Sequential()
```

**Creating Network**

```python
ann.add(Dense( 4, activation='relu'))
ann.add(Dense(10, activation='relu' ))
ann.add(Dense(10, activation='relu' ))
ann.add(Dense( 3, activation='softmax'))
```

```python
ann.compile(optimizer='adam', loss= 'categorical_crossentropy', metrics=['accuracy'])
```

```python
ann.fit(X_train, y_train, batch_size=4, epochs=30)
```

```python
ann.summary()
```

```python
y_pred= ann.predict(X_test)
```

**we use argmax to select the max value in column**

```python
y_pred= np.argmax(y_pred, axis=1)
```

```python
y_test
```

```python
y_test= np.argmax(y_test, axis=1)
```

```python
y_test
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```
